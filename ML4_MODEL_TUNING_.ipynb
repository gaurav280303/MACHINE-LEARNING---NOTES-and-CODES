```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('iris')
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('species', axis=1)
y=df['species']
```

```python
x
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn=KNeighborsClassifier(n_neighbors=57) #We can change this n_neighbor like 3,5 ,25, and more. Each time I might get different accuracy
model_knn.fit(x_train, y_train)
```

```python
model_knn.score(x_test, y_test)
```

If you are getting accuracy score 100 % , model is overfit

SVM

```python
from sklearn.svm import SVC
```

```python
model_svm=SVC(C=30, kernel='rbf', gamma='auto')  #C is a punishment/penalty, agar decision boundary ke andar koi aaya
model_svm.fit(x_train, y_train)
```

```python
model_svm.score(x_test, y_test)
```

```python
#Applying cross validation in SVM model
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model_svm, x,y, cv=6, scoring='accuracy')
scores
```

Generally Each time we tweak/change parameters in models, we might get different accuracy. So how to identify which combination(grid) to use to get maximum accuracy, for that we use GRID SEARCH CV

GRID SEARCH CV Code Implementation

```python
from sklearn.model_selection import GridSearchCV
```

Performing HYPERPARAMETERTUNING through GRID Search CV

Grid Search CV for SVM Model

```python
classifier=GridSearchCV((model_svm),{
    'C': [1,10,20,30,50],
    'kernel': ['rbf', 'linear', 'poly'],
    'gamma': ['auto', 'scale']
}, cv=5, return_train_score=False) #Inside the dictionary there are multiple hyperparameters of the SVM model only
```

```python
classifier.fit(x,y)
```

```python
results_gridcv_svm=pd.DataFrame(classifier.cv_results_)
results_gridcv_svm

#This will show the different cobinations(grid) of the hyperparameters ,
#which you gave while creating SVM model in Grid Search CV inside the dictionary(in above code)
```

```python
results_gridcv_svm.columns
```

```python
#The above output result is too big , but I am only interested in 'param_C', 'param_gamma', 'param_kernel, 'mean_test_score' i.e, accuracy

results_gridcv_svm[['param_C', 'param_gamma', 'param_kernel', 'mean_test_score']]
```

What Actually Happens i n RANDOM SEARCH CV model:
Randomly selects 10 parameter combinations from the 40 possible ones

For each of these 10 selected combinations, it does 5-fold CV

So it trains: 10 combinations × 5 folds = 50 models total

Random Selection: Randomly picks 10 hyperparameter combinations from all possibilities

Cross-Validation: For each combination, performs 5-fold CV

Scoring: Evaluates performance on validation folds

Best Model: Returns the combination with highest average validation score

Missing Parameters (Usually Default)
Scoring metric: Not specified, uses classifier's default (usually accuracy for classification)

Random state: Not set, so results may vary between runs

n_jobs: Not specified, uses single processor

Advantages Over GridSearchCV
Faster: Tests only a subset of parameters

Better for large spaces: More efficient when many hyperparameters

Can find good combinations: Often finds near-optimal parameters quickly

This approach balances computational efficiency with finding good hyperparameter values for your KNN model.

What is return_train_score? (Simple Explanation)
Think of it like this: When you're practicing for an exam...

Training score = Your score on practice questions (the ones you studied)

Test/Validation score = Your score on real exam questions (new, unseen questions)

The Setting:
python
return_train_score=False
What This Means:
return_train_score=False = "Don't show me my practice scores, just show me my exam scores"

Simple Example:
Let's say you're tuning your KNN model with different numbers of neighbors:

n_neighbors	Practice Score	Exam Score
1	98%	75%
3	95%	82%
5	92%	85%
With return_train_score=False, the code only remembers:

Best model: n_neighbors=5 with 85% exam score

Why Turn It Off?
Saves Memory: Practice scores take up space to remember

Faster: Less data to process and store

Focus on What Matters: Only care about how well you'll do on NEW questions (the exam), not the ones you already practiced

When You Might Want It On:
python
return_train_score=True  # Show me both practice AND exam scores
Use this when you want to check for overfitting:

If practice score = 98% but exam score = 75% → You memorized instead of learning!

This helps you understand if your model is too complex

Bottom Line:
return_train_score=False = "Just tell me which settings work best on new data, don't bother remembering how they did on the training data."

It's like saying: "I don't care how well I memorized the practice questions - I only care about how I'll perform on the actual test!"
