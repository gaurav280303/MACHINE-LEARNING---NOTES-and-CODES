ENSEMBLE LEARNING

1. STACKING

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import StackingClassifier
import seaborn as sns
```

```python
df=sns.load_dataset("iris")
df
```

```python
import pandas as pd
x=df.drop('species', axis=1)
y=df['species']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
y_encoded=le.fit_transform(y)
y_encoded
```

```python
x_train,x_test, y_train, y_test=train_test_split(x, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)
```

```python
base_learner_models={
    'LogisticRegression': LogisticRegression(max_iter=1000),
    'DecisionTreeClassifier': DecisionTreeClassifier( random_state=42),
    'SVC': SVC(probability=True, kernel='rbf', random_state=42)
}
```

```python
meta_model=KNeighborsClassifier()
```

STACKING

```python
stacking_classifier=StackingClassifier(
    estimators=list(base_learner_models.items()),
    final_estimator=meta_model,
    cv=5 #kfold =5 , cross validation
)
```

```python
stacking_classifier.fit(x_train, y_train)
```

```python
y_pred=stacking_classifier.predict(x_test)
y_pred
```

```python
accuracy=accuracy_score(y_test, y_pred)
accuracy
```

```python
confusion_matrix(y_test, y_pred)
```

2. BAGGING---
Bagging converts overfit(low biad high variace)  models in to general fit

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier(
    n_estimators=100, #Number of decision trees we want to make
    max_depth=None, #IF its none, means it will go till the purest form of leaf nodes
    max_features='sqrt', #If there are 4 features(columns) in input data it will do sqrt , meanssqrt(4) =2, means randomly 2 features selected among 4 for creating next subsequent nodes in decision trees
    random_state=42,
    bootstrap=True #Whether bootstrap samples are used when building trees. If False, the whole dataset is used to build each tree.
)
```

```python
rf_model.fit(x_train, y_train)
```

```python
y_pred_rf=rf_model.predict(x_test)
y_pred_rf
```

```python
accuracy=accuracy_score(y_test, y_pred_rf)
accuracy
```

```python
confusion_matrix(y_test, y_pred_rf)
```

```python
print(classification_report(y_test, y_pred_rf))
```

3. BOOSTING---
Boosting converts underfit model(high bias & high variance) into general fit .

ADABOOST

```python
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
```

```python
adaboost_model= AdaBoostClassifier(
    n_estimators=100, #100 decision trees
    learning_rate=1.0,
    random_state=42
)
```

With learning_rate=1.0:

text
Tree1: "The answer is A!" ← We believe it 100%
Tree2: "No, the answer is B!" ← We believe it 100%  
Tree3: "Actually, it's C!" ← We believe it 100%
→ Model gets confused, changes mind too much

With learning_rate=0.1:

text
Tree1: "The answer is A!" ← We believe it 10%
Tree2: "No, the answer is B!" ← We believe it 10%
Tree3: "Actually, it's C!" ← We believe it 10%
→ Model combines all opinions slowly and carefully

Simple Rule:
learning_rate=1.0 → Fast but risky

learning_rate=0.1 → Slow but safe

learning_rate=0.5 → Balanced approach

```python
adaboost_model.fit(x_train, y_train)
```

```python
y_pred_adaboost=adaboost_model.predict(x_test)
y_pred_adaboost
```

```python
accuracy_score(y_test, y_pred_adaboost)
```

```python
confusion_matrix(y_test, y_pred_adaboost)
```

```python
print(classification_report(y_test, y_pred_adaboost))
```
