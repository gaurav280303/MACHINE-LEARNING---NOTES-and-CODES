1. LOGISTIC REGRESSION

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df.drop(['deck','embark_town', 'parch', 'who', 'embark_town'],axis=1,inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df['age'].fillna(df['age'].mean(),inplace=True)
```

```python
df.dropna(subset='embarked',inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df
```

```python
df['sex']=le.fit_transform(df['sex'])
```

```python
df['embarked']=le.fit_transform(df['embarked'])
df['class']=le.fit_transform(df['class'])
df['adult_male']=le.fit_transform(df['adult_male'])
df['alive']=le.fit_transform(df['alive'])
df['alone']=le.fit_transform(df['alone'])
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
df['age']=sc.fit_transform(df[['age']])
df['fare']=sc.fit_transform(df[['fare']])
df['pclass']=sc.fit_transform(df[['pclass']])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('survived',axis=1)
y=df['survived']
```

```python
x.drop(['alive', 'alone', 'adult_male'], axis=1, inplace=True)
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_pred, y_test)
```

```python
confusion_matrix(y_test, y_pred, labels=[1,0])
```

```python
print(classification_report(y_test, y_pred))
```
