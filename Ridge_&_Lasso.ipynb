Normalization vs Standardization
Both are techniques to scale your features, but they do it differently.

Normalization (Min-Max Scaling)
What it does: Squeezes data into a fixed range, usually [0, 1]

Formula: (x - min) / (max - min)

Effect: All values end up between 0 and 1

Example:
Let's say we have house sizes: [1000, 2000, 3000, 4000] sq ft

Normalized:

1000 → (1000-1000)/(4000-1000) = 0

2000 → (2000-1000)/(4000-1000) = 0.33

3000 → 0.67

4000 → 1.0

Result: [0, 0.33, 0.67, 1.0]

Standardization (Z-score Normalization)
What it does: Centers data around mean=0 with standard deviation=1

Formula: (x - mean) / standard_deviation

Effect: Data has mean=0, std=1, but no fixed range

Same example:
House sizes: [1000, 2000, 3000, 4000]

Mean = 2500

Standard deviation ≈ 1291

Standardized:

1000 → (1000-2500)/1291 ≈ -1.16

2000 → (2000-2500)/1291 ≈ -0.39

3000 → 0.39

4000 → 1.16

Result: ≈ [-1.16, -0.39, 0.39, 1.16]

RIDGE REGRESSION

Part 1: The Problem with Linear Regression (The "Know-It-All" Model)
Imagine you are trying to predict how tall a sunflower will grow based on two things:

The amount of sunlight it gets.

The amount of water it gets.

You plant 10 sunflowers and carefully measure everything. A Linear Regression model is like a smart student who looks at your data and draws a straight line (or a flat plane) that best fits all your points.

This seems great! So what's the problem?

The problem is that this smart student is a know-it-all. It tries to be 100% perfect for your exact 10 sunflowers. It will twist and turn its formula to pass perfectly through every single data point, even the weird ones.

Example of the Problem:
Let's say one of your sunflowers was chewed by a bug and grew to be very short, even with lots of sun and water. The know-it-all student sees this and says, "Aha! I see a pattern! If a sunflower gets exactly 5 hours of sun and 2 cups of water, it MUST be short because of this one time." It learns this "lesson" too well.

The Drawback: Overfitting
This is the main drawback. Overfitting means the model has learned the training data too well, including all the noise and random quirks. When you show it new data (e.g., your 11th sunflower), it performs poorly because it's applying the weird, overly-specific rules it learned from your first 10.

In Technical Terms:
Linear regression models can have very high variance. Their parameters (the numbers they multiply sunlight and water by) can become extremely large and sensitive to the training data. They model the noise, not just the underlying signal.

Part 2: The Heroes - Ridge and Lasso Regression (The "Well-Rounded" Students)
To fix the know-it-all, we hire two tutors: Ridge and Lasso. Their job is to tell the student, "It's okay to be a little bit wrong on the homework if it means you'll do better on the final exam."

They both do the same fundamental thing: They punish the model for having huge numbers in its formula.

The Core Idea: Regularization
This "punishment" is called Regularization. It's a penalty we add to the model's goal.

Linear Regression's Goal: "Make the prediction error as small as possible."

Ridge & Lasso's Goal: "Make the prediction error as small as possible, AND keep the model's numbers (coefficients) from getting too big."

This forces the model to be simpler and less focused on the tiny details.

Part 3: The Formulas - How the Punishment Works
First, let's understand the goal without punishment. Linear Regression tries to minimize the Sum of Squared Errors (SSE).

SSE = Σ (Actual Value - Predicted Value)²
(We sum up all the squared mistakes.)

Now, let's add the punishment!

1. Ridge Regression (The "Gentle" Tutor)
Ridge says, "I don't care which numbers are big, I just want all of them to be reasonably small." It punishes the square of the coefficients.

The Ridge Formula:
Minimize [ SSE + λ * Σ (Coefficients)² ]

λ (lambda): This is the "Punishment Knob." If we turn λ to zero, there's no punishment, and Ridge acts just like Linear Regression. If we turn λ way up, the model gets so scared of having large coefficients that they all get pushed towards zero, making a very simple (but potentially underfitting) model.

Σ (Coefficients)²: This is the "L2 Penalty." It's the sum of the squares of all the coefficients (except the intercept).

What it does: Ridge shrinks coefficients towards zero, but they will rarely become exactly zero. All features stay in the model, but their influence is reduced.

2. Lasso Regression (The "Strict" Tutor)
Lasso says, "Not only do I want the numbers to be small, but I also want to figure out which features are actually important." It punishes the absolute value of the coefficients.

The Lasso Formula:
Minimize [ SSE + λ * Σ |Coefficients| ]

λ (lambda): The same "Punishment Knob."

Σ |Coefficients|: This is the "L1 Penalty." It's the sum of the absolute values of all the coefficients.

What it does: Lasso also shrinks coefficients, but it has a special power: it can force coefficients all the way to zero. This is like a built-in feature selector.

Part 4: The Key Difference & A Perfect Example
Let's add a third, silly feature to our sunflower prediction:
3. The number of clouds in the sky on the day the gardener sneezed. (This is obviously a useless feature!).

We train all three models:

Linear Regression (Know-It-All): It will use all three features. It will assign some number (coefficient) to the "clouds/sneeze" feature, no matter how meaningless, to try and fit the training data perfectly. This makes the model worse at predicting new sunflowers.

Ridge Regression (Gentle Tutor): It will shrink the coefficients for all three features. The coefficient for "clouds/sneeze" will become very small, but it will not be zero. The feature is still technically in the model.

Lasso Regression (Strict Tutor): It will likely shrink the coefficient for "clouds/sneeze" all the way to ZERO. It effectively kicks that useless feature out of the model! This makes the model simpler and easier to interpret.

The Key Difference Summarized:

Ridge is better when you have many small/medium effects and you believe all features are relevant.

Lasso is better when you suspect only a few features are truly important and you want to automatically identify them.

Summary Table
Feature	Linear Regression	Ridge Regression	Lasso Regression
Main Goal	Fit the data as closely as possible	Fit data well with small coefficients	Fit data well and select important features
Handles Overfitting?	No	Yes	Yes
Penalty Term	None	L2 Penalty: Σ(Coeff)²	L1 Penalty: `Σ	Coeff	`
What happens to coefficients?	Can become very large	Shrunk towards zero, but not zero	Shrunk towards zero, and can become exactly zero
Best for...	Simple, well-behaved data	When you have many relevant features	When you have many features and want to find the most important ones
Analogy	The Know-It-All Student	The Gentle Tutor	The Strict Tutor
So, in short, we use Ridge and Lasso to prevent overfitting by penalizing complexity, which leads to models that generalize much better to new, unseen data. Lasso gives us the bonus of automatic feature selection.

Step 3: Pick the Best One
We choose the Lambda that works best on the validation set (not the training set)!

Why did we do values flatten here , and what does the graph tells
Great questions! y_test was a pandas DataFrame with one column, and y_pred_ridge is a NumPy array. To subtract them element by element and get the difference (which are the residuals), they need to have compatible shapes. .values extracts the data from the pandas DataFrame as a NumPy array, and .flatten() converts that array into a 1-dimensional array, matching the shape of y_pred_ridge.

The graph you see is a distribution plot of the residuals. Residuals are the differences between your model's predicted values (y_pred_ridge) and the actual values (y_test).

What the graph tells you:

Centered around zero: Ideally, for a good model, the distribution of residuals should be centered around zero. This means that, on average, your model's predictions are neither consistently overestimating nor underestimating the actual values.
Shape of the distribution: The shape tells you about the spread of the errors. A narrow, bell-shaped distribution (like the one you have, which looks somewhat like a normal distribution) is generally desirable. It suggests that most of your predictions are relatively close to the actual values, and larger errors are less frequent.
In short, this graph helps you visually assess how well your model's predictions align with the actual data. A distribution of residuals centered at zero with a relatively small spread indicates a better-performing model.
